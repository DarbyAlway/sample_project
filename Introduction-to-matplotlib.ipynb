# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline

# 1. prepare data
x= [1,2,3,4]
y = [11,22,33,44]

# 2. setup plot
fig,ax = plt.subplots(figsize=(10,10)) #  width and height

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = 'Simple Plot',
       xlabel = 'x-axis',
       ylabel = 'y-axis')

# 5. Save and show(you save the whole figure)
fig.savefig('sample-plot.png')
```

# Two options for subplots

```python
# subplot option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows = 3,
                                           ncols = 2,
                                           figsize = (10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nutt_butter_prices.keys(),nutt_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig,ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10,5))
# Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nutt_butter_prices.keys(),nutt_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? ( pyplot or matplotlib oo method ?)
* When plotting something quickly, okay to use the pyplot method
* When plotting something more advance, use the OO method

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
## OO method mix with pyplot method
import matplotlib.pyplot as plt
fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind = 'scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
ax.set_xlim([45,100]);
```

```python
#Set the style
plt.style.use('whitegrid')
# OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50["chol"],
                     c=over_50['target'],
                    cmap = "winter");
#Customize
ax.set(title="Heart Disease and Cholesterol levels",
       xlabel="Age",
       ylabel="Choresterol");

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
#style the plot
# Subplot of chol, age ,thalach
fig, (ax0,ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize=(10,10),
                             sharex=True)
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50['chol'],
                      c=over_50['target'])

#customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel='Cholesterol')

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a meanline 
ax0.axhline(y=over_50["chol"].mean(),
            linestyle = "--");

#Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50['target']);
#Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="thalach");

# #Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

#Add a mean line 
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16,fontweight = "bold")
```
